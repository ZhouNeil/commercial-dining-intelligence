```python
import ast
philly_data['hours'] = philly_data['hours'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
def extract_hours_features(hours_dict):
    if not isinstance(hours_dict, dict) or len(hours_dict) == 0:
        return pd.Series([0, 0, 0])

    is_open_morning = 0
    is_open_latenight = 0
    open_on_weekends = 0

    for day, hours in hours_dict.items():
        try:
            open_time, close_time = hours.split('-')

            # Convert to hour integers
            open_hour = int(open_time.split(':')[0])
            close_hour = int(close_time.split(':')[0])

            # Morning check
            if open_hour < 10:
                is_open_morning = 1

            # Late night check
            if close_hour >= 22:
                is_open_latenight = 1

            # Weekend check
            if day in ['Saturday', 'Sunday']:
                open_on_weekends = 1

        except:
            continue

    return pd.Series([is_open_morning, is_open_latenight, open_on_weekends])
```

```python
philly_data[['time_is_open_morning', 'time_is_open_latenight', 'time_open_on_weekends']] = philly_data['hours'].apply(extract_hours_features)
```

```python
philly_data[['time_is_open_morning', 'time_is_open_latenight', 'time_open_on_weekends']] = philly_data[['time_is_open_morning', 'time_is_open_latenight', 'time_open_on_weekends']].fillna(0).astype(int)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Make a safe copy
philly_data = philly_data.copy()

# Convert comma-separated strings into lists
philly_data['categories'] = philly_data['categories'].apply(
    lambda x: [item.strip() for item in x.split(',')] if isinstance(x, str) else []
)

# One-hot encode
mlb = MultiLabelBinarizer()

categories_df = pd.DataFrame(
    mlb.fit_transform(philly_data['categories']),
    columns=mlb.classes_,
    index=philly_data.index
)

# ✅ Add prefix "cat_" to all category columns
categories_df.columns = ['cat_' + col.lower().replace(' ', '_') for col in categories_df.columns]

# Merge back
philly_data = pd.concat([philly_data, categories_df], axis=1)
```

```python
import re

# Make safe copy
philly_data = philly_data.copy()

# -----------------------------
# Step 1: Parse attributes column into dict
# -----------------------------
def safe_parse(x):
    if isinstance(x, dict):
        return x
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return {}
    return {}

philly_data['attributes'] = philly_data['attributes'].apply(safe_parse)

# -----------------------------
# Step 2: Flatten + clean values
# -----------------------------
def clean_value(val):
    if isinstance(val, str):
        val = val.strip()

        # Handle nested dict stored as string
        if val.startswith('{') and val.endswith('}'):
            try:
                return ast.literal_eval(val)
            except:
                return {}

        # Normalize unicode-like strings: u'free'
        val = val.replace("u'", "").replace("'", "")

        # Boolean conversion
        if val.lower() == 'true':
            return 1
        if val.lower() == 'false':
            return 0

        # Numeric conversion
        if val.isdigit():
            return int(val)

        return val

    return val


def flatten_attributes(attr_dict):
    result = {}

    if not isinstance(attr_dict, dict):
        return result

    for key, val in attr_dict.items():
        val = clean_value(val)

        # If nested dictionary → flatten
        if isinstance(val, dict):
            for sub_key, sub_val in val.items():
                sub_val = clean_value(sub_val)

                col_name = f"attr_{key}_{sub_key}"
                result[col_name] = 1 if sub_val in [True, 1] else 0

        else:
            col_name = f"attr_{key}"

            # Boolean-like
            if val in [True, 1]:
                result[col_name] = 1
            elif val in [False, 0]:
                result[col_name] = 0
            else:
                result[col_name] = val

    return result

# Apply flattening
attr_df = philly_data['attributes'].apply(flatten_attributes).apply(pd.Series)

# -----------------------------
# Step 3: Clean column names
# -----------------------------
attr_df.columns = [
    re.sub(r'[^a-z0-9]+', '_', col.lower()).strip('_')
    for col in attr_df.columns
]

# -----------------------------
# Step 4: Fill missing values
# -----------------------------
attr_df = attr_df.fillna(0)

# -----------------------------
# Step 5: Merge back
# -----------------------------
philly_data = pd.concat([philly_data, attr_df], axis=1)

# Optional: drop original column
# philly_data = philly_data.drop(columns=['attributes'])
```
